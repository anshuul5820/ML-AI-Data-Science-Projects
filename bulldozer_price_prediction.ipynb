```python
X_train,y_train=df_train.drop("SalePrice",axis=1),df_train.SalePrice
X_val,y_val=df_val.drop("SalePrice",axis=1),df_val.SalePrice
```

```python
X_train.shape,y_train.shape,X_val.shape,y_val.shape
```

```python
from sklearn.metrics import mean_squared_log_error, mean_absolute_error,r2_score

def rmsle(y_test,y_preds):
    """
    This function finds out Root Mean Square Log Error
    """
    return np.sqrt(mean_squared_log_error(y_test,y_preds))
    
def show_scores(model):
    
    train_preds=model.predict(X_train)
    val_preds=model.predict(X_val)
    
    scores={"Training MAE":mean_absolute_error(y_train,train_preds),
           "Valid MAE":mean_absolute_error(y_val,val_preds),
           "Training RMSLE":rmsle(y_train,train_preds),
           "Valid RMSLE":rmsle(y_val,val_preds),
           "Training R^2 score":r2_score(y_train,train_preds),
           "Valid R^2":r2_score(y_val,val_preds)}
    return scores

    
```

## The data needs to be preprocessed before testing, since the demo data has 102 columns & test data has 52 columns.

```python
%%time
df_test1=pd.read_csv("/content/drive/My Drive/Data for Colab notebooks/Bulldozer price prediction dataset/Test.csv",low_memory=False,parse_dates=["saledate"])

def preprocess_data(df_test):
    
    """
    This function converts non-integral data into categorical data
    This also fills empty integer columns with median
    """
    df_test["saleYear"] = df_test.saledate.dt.year
    df_test["saleMonth"] = df_test.saledate.dt.month
    df_test["saleDay"] = df_test.saledate.dt.day
    df_test["saleDayofweek"] = df_test.saledate.dt.dayofweek
    df_test["saleDayofyear"] = df_test.saledate.dt.dayofyear

    # Drop original saledate
    df_test.drop("saledate", axis=1, inplace=True)
    
    for key,values in df_test.items():
        if pd.api.types.is_numeric_dtype(values):
            if pd.isnull(values).sum():
                df_test[key+"_is_missing"]=pd.isnull(values)
                df_test[key]=values.fillna(values.median())
            
        if not pd.api.types.is_numeric_dtype(values):
                df_test[key+"_is_missing"]=pd.isnull(values)
                df_test[key]=pd.Categorical(values).codes+1
            
    return df_test
```

```python
import seaborn as sns

def plot_features(columns,importances,n=20):
    df=(pd.DataFrame({"features":columns,
                     "feature_importances":importances})
                .sort_values("feature_importances",ascending=False)
                   .reset_index(drop=True))
    plt.figure(figsize=(20,10))
    fig,ax=plt.subplots()
    ax.barh(df["features"][:n],df["feature_importances"][:20])
    ax.set_ylabel("features")
    ax.set_xlabel("feature importances")
    
    sns.barplot(x="feature_importances",
                y="features",
                data=df[:n],
                orient="h")
    
```
